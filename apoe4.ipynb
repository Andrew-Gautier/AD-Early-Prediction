# Checking percentage of new dataset IDs have APOE4 genetic info.  

```python
import pandas as pd

df = pd.read_csv('investigator_ftldlbd_nacc72.csv', low_memory=False)
```

```python

# Unique subjects and their NACCAPOE values
unique_subjects = df.drop_duplicates(subset='NACCID')[['NACCID', 'NACCAPOE']]

total_unique = len(unique_subjects)
has_apoe = unique_subjects[unique_subjects['NACCAPOE'] != 9]
missing_apoe = unique_subjects[unique_subjects['NACCAPOE'] == 9]

print(f"Total unique subjects:              {total_unique}")
print(f"Subjects with APOE data (≠9):       {len(has_apoe)}  ({len(has_apoe)/total_unique*100:.1f}%)")
print(f"Subjects missing APOE data (=9):    {len(missing_apoe)}  ({len(missing_apoe)/total_unique*100:.1f}%)")

# NACCAPOE value mapping per data dictionary
apoe_labels = {
    1: 'e3,e3',
    2: 'e3,e4',
    3: 'e3,e2',
    4: 'e4,e4',
    5: 'e4,e2',
    6: 'e2,e2',
    9: 'Missing/Unknown',
}

print("\nBreakdown by NACCAPOE value (unique subjects):")
counts = unique_subjects['NACCAPOE'].value_counts().sort_index()
for val, count in counts.items():
    label = apoe_labels.get(val, 'Unknown code')
    print(f"  {val} ({label}): {count}  ({count/total_unique*100:.1f}%)")
```
